# PEMODELAN PENDULUM

Pendulum sederhana merupakan sistem mekanik yang terdiri dari massa yang tergantung pada tali atau batang kaku yang berayun di sekitar titik tetap. Pemodelan matematis pendulum sederhana dapat dilakukan dengan menerapkan hukum kedua Newton dalam bentuk gerak rotasi.

## Persamaan Dinamika Pendulum
Berdasarkan gambar, gaya yang bekerja pada pendulum terdiri dari:

*   Gaya gravitasi ($𝑚𝑔$) yang memiliki komponen $mgsin\theta$ yang menyebabkan gerakan rotasi
*   Torsi eksternal ($T_e$) yang dapat diterapkan pada sistem.


Dari hukum kedua Newton dalam rotasi, total torsi ($T$) yang bekerja pada sistem adalah:
$$T = J\alpha$$
dengan:\
$T$ adalah total torsi ukuran gaya yang menyebabkan benda berputar terhadap suatu titik atau sumbu tertentu dalam (Newton meter, $Nm$)\
$J$ adalah momen inersia sistem, adalah ukuran resistansi suatu benda terhadap perubahan gerak rotasi di sekitar suatu sumbu. $kgm^2$\
$\alpha$ adalah percepatan sudut adalah perubahan kecepatan sudut ($\omega$)terhadap waktu. dalam $rad/s^2$

## Momen Inersia
Momen inersia dengan massa $m$ yang terletak pada ujung batang panjang $l$ adalah:
$$J=ml^2$$
Torsi akibat gaya gravitasi diberikan oleh:
$$T_g=-mglsin\theta$$

Sehingga Persamaan Pendulum menjadi:

$$ml^2{\alpha}=-mglsin\theta+T_e$$

atau:

$$\alpha=-\frac{g}{l}sin\theta+\frac{T_e}{ml^2}$$

## Linearisasi Sistem Pendulum

Untuk analisis kestabilan atau kontrol sistem, seringkali digunakan pendekatan linier. **Jika sudut kecil** ($sin\theta≈\theta$)

Persamaan menjadi:
$$\alpha≈-\frac{g}{l}\theta+\frac{T_e}{ml^2}$$

Rubah ke persamaan diferensial percepatan ($\alpha$) adalah turunan kedua $\theta$ sehingga:

$$\ddot{\theta}+\frac{g}{l}\theta=\frac{T_e}{ml^2}$$

## State Space Model

Definisi variabel state space
Untuk mengubah sistem orde dua menjadi orde satu, kita mendefinisikan dua variabel keadaan:

$$x_1=\theta$$
$$x_2=\dot{\theta}$$

Dengan demikian, turunan dari masing-masing variabel keadaan adalah:
$$\dot{x_1}=x_2$$
$$\dot{x_2}=\ddot{\theta}$$

sehingga:

$$\begin{bmatrix}
\dot{x_1}\\
\dot{x_2}
\end{bmatrix}=
\begin{bmatrix}
0 & 1\\
-\frac{g}{l} &0
\end{bmatrix}
\begin{bmatrix}
{x_1}\\
{x_2}
\end{bmatrix}
+
\begin{bmatrix}
0\\
\frac{1}{ml^2}
\end{bmatrix}
T_e
$$
